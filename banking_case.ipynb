```python
bins=[0, 100000, 300000, float('inf')]
labels=['Low', 'Mid', 'High']

df['Income Band']=pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)

df['Income Band'].value_counts().plot(kind='bar')
```

```python
#Examine the distribution of unique categories in categorical columns
categorical_cols=df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation","Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
    print(f"Value counts for '{col}':")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation","Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation","Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="GenderId")
```

```python
# Plot countplots for all categorical columns except "Occupation"
for col in categorical_cols:
    if col == "Occupation":
        continue

    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col)
    plt.title(f"Countplot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)  # Rotate labels for readability if needed
    plt.tight_layout()
    plt.show()
```

```python
#Histplot of value counts for different occupation

for col in  categorical_cols:
    if col=="Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Ocupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

NUMERICAL ANALYSIS

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Bank Deposits', 'Checking Accounts',
    'Saving Accounts', 'Foreign Currency Account', 'Business Lending'
]

# Univariate analysis and visualization
plt.figure(figsize=(15, 25))  # Increased height to fit 9 plots
for i, col in enumerate(numerical_cols):
    plt.subplot(9, 1, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Bank Deposits', 'Checking Accounts',
    'Saving Accounts', 'Foreign Currency Account', 'Business Lending'
]
correlation_matrix=df[numerical_cols].corr()
# Univariate analysis and visualization
plt.figure(figsize=(5,5))  # Increased height to fit 9 plots
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

1. Strong Relationship Between Bank Deposits and Checking Accounts
Correlation: 0.84

This high positive correlation suggests that customers with higher checking account balances tend to also have high bank deposits.

This may indicate similar financial behavior or overlapping account features.

📌 2. Bank Deposits and Saving Accounts Are Closely Linked
Correlation: 0.75

Customers who hold savings accounts often also maintain significant bank deposits.

This is expected behavior in personal finance, reflecting conservative or saving-focused users.

📌 3. Bank Loans Have Moderate Positive Correlation with Business Lending
Correlation: 0.42

Individuals or businesses taking personal bank loans are also likely to be involved in business lending.

This could indicate entrepreneurial users with multiple credit products.

📌 4. Estimated Income Shows Weak to Moderate Correlation with Other Variables
Highest with Superannuation Savings (0.37) and Bank Loans/Business Lending (~0.33)

Higher income individuals are moderately likely to invest more in retirement savings and take bank loans (perhaps due to higher eligibility).

However, no very strong linear correlation suggests income alone does not determine financial product behavior.

📌 5. Credit Card Balance Has Modest Correlation with Foreign Currency Account (0.36) and Bank Deposits (0.38)
This could imply that individuals with higher credit card usage are also actively managing international or savings-related accounts.

Possibly reflects a financially active or globally involved customer segment.
